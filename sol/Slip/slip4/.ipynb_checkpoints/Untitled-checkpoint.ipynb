```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers
np.random.seed(42)  # Set a seed for reproducibility
data = np.random.randint(1, 100, 50)

# Create a figure and set of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Line chart
axs[0, 0].plot(data, color='blue', marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('Data Point Index')
axs[0, 0].set_ylabel('Value')

# Scatter plot
axs[0, 1].scatter(range(1, 51), data, color='green', marker='x')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('Data Point Index')
axs[0, 1].set_ylabel('Value')

# Histogram
axs[1, 0].hist(data, bins=10, color='purple', alpha=0.7)
axs[1, 0].set_title('Histogram')
axs[1, 0].set_xlabel('Value')
axs[1, 0].set_ylabel('Frequency')

# Box plot
axs[1, 1].boxplot(data, vert=False, patch_artist=True)
axs[1, 1].set_title('Box Plot')
axs[1, 1].set_xlabel('Value')

# Add some styling
for ax in axs.flat:
    ax.grid(True)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
