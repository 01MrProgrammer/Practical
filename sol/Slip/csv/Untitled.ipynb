```python
import pandas as pd
data = pd.read_csv("Data.csv")
mean_salary = data['Salary'].mean()
mean_age = data['Age'].mean()
data['Salary'].fillna(mean_salary, inplace=True)
data['Age'].fillna(mean_age, inplace=True)

print(data)
```

```python
import matplotlib.pyplot as plt
names = ['Rushi', 'sachin', 'vinit', 'uday', 'nitin']
salaries = [50000, 60000, 75000, 55000, 80000]
plt.figure(figsize=(10, 6))
plt.plot(names, salaries, marker='o', linestyle='-', color='b')
plt.title('Names vs. Salary')
plt.xlabel('Names')
plt.ylabel('Salary')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
