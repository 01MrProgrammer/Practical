```python
#s13
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_data = pd.read_csv(r"iris.csv",names=["sepallenghth","sepalwidth","petallength",
                               "petalwidth","Species"])
print(iris_data)
# Extract petal length and petal width columns
petal_length = iris_data['petallength']
petal_width = iris_data['petalwidth']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(petal_length, petal_width, c='blue', marker='o')
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)

plt.show()
```
