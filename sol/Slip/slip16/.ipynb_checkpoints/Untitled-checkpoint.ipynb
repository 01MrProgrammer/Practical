```python
import numpy as np
import matplotlib.pyplot as plt
subjects = ['Math', 'Science', 'English', 'History', 'Geography']
marks = [85, 90, 78, 92, 88]

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title('Marks in Subjects (Pie Chart)')
plt.show()

# Bar chart
plt.figure(figsize=(10, 5))
plt.bar(subjects, marks, color=['red', 'green', 'blue', 'yellow', 'purple'])
plt.title('Marks in Subjects (Bar Chart)')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.show()
```
