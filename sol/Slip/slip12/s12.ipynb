```python
#s12
import matplotlib.pyplot as plt
import numpy as np

# Generate a random array of 50 integers
random_data = np.random.randint(1, 100, 50)

# Create a figure with multiple subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Data Visualization', fontsize=16)

# Line Chart
axes[0, 0].plot(random_data, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6)
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('Data Points')
axes[0, 0].set_ylabel('Values')
axes[0, 0].grid(True)

# Scatter Plot
axes[0, 1].scatter(range(1, 51), random_data, color='red', marker='^')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('Data Points')
axes[0, 1].set_ylabel('Values')
axes[0, 1].grid(True)

# Histogram
axes[1, 0].hist(random_data, bins=10, color='green', alpha=0.7)
axes[1, 0].set_title('Histogram')
axes[1, 0].set_xlabel('Values')
axes[1, 0].set_ylabel('Frequency')

# Box Plot
axes[1, 1].boxplot(random_data, vert=False, patch_artist=True)
axes[1, 1].set_title('Box Plot')
axes[1, 1].set_xlabel('Values')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
import random
import numpy as np

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Salary': [60000, 70000, 55000, None, 65000],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'Finance']
}

df = pd.DataFrame(data)

# Add 10 rows with some missing and duplicate values
for _ in range(10):
    row = {
        'Name': random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Eva']),
        'Salary': random.choice([60000, 70000, 55000, None]),
        'Department': random.choice(['HR', 'Finance', None])
    }
    df = df.append(row, ignore_index=True)

# Drop all null and empty values
df = df.dropna()

# Print the modified data frame
print("Modified Data Frame:")
print(df)
```
