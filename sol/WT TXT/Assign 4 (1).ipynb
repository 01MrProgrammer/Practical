```python
#Set A 1)
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data = np.random.randint(1, 100, 50)

# Line chart
plt.figure(figsize=(10, 5))
plt.plot(data, color='blue', marker='o', linestyle='-')
plt.title('Line Chart')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python

# Scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(range(len(data)), data, color='green', marker='o')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=10, color='orange', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 5))
plt.boxplot(data, vert=False, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

```python
#2)
# Add outliers
data_with_outliers = np.append(data, [150, 200])

# Box plot with outliers
plt.figure(figsize=(10, 5))
plt.boxplot(data_with_outliers, vert=False, patch_artist=True)
plt.title('Box Plot with Outliers')
plt.xlabel('Value')

plt.show()
```

```python
#3)
import matplotlib.pyplot as plt

subjects = ["Math", "Science", "History", "English"]
marks = [85, 90, 78, 88]

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(marks, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.title('Marks by Subject (Pie Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python

# Bar chart
plt.figure(figsize=(10, 5))
plt.bar(subjects, marks, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks by Subject (Bar Chart)')
plt.show()
```

```python
#6)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from a CSV file
iris_data = pd.read_csv(r"Iris.csv")

# Create histograms for each species
plt.figure(figsize=(12, 6))
sns.histplot(data=iris_data, x='SepalLengthCm', hue='Species', kde=True, element='step', common_norm=False)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.title('Histogram of Sepal Length by Species')
plt.legend(title='Species')

plt.show()
```

```python
#set c 1)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from a CSV file
iris_data = pd.read_csv(r"Iris.csv")

# Task 1: Create a pairplot and check which flower species seems to be the most separable
sns.pairplot(iris_data, hue='Species')
plt.title("Pairplot of Iris Dataset")
plt.show()
```

```python
# 2)
# Sepal Length
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_data)
plt.title("Box Plot of Sepal Length by Species")
plt.show()

# Sepal Width
plt.figure(figsize=(12, 8))
sns.boxplot(x="Species", y="SepalWidthCm", data=iris_data)
plt.title("Box Plot of Sepal Width by Species")
plt.show()

# Petal Length
plt.figure(figsize=(12, 8))
sns.boxplot(x="Species", y="PetalLengthCm", data=iris_data)
plt.title("Box Plot of Petal Length by Species")
plt.show()

# Petal Width
plt.figure(figsize=(12, 8))
sns.boxplot(x="Species", y="PetalWidthCm", data=iris_data)
plt.title("Box Plot of Petal Width by Species")
plt.show()
```

```python
#3)
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris_data, kind="kde")
plt.show()
```
