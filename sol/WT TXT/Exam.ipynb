```python
#S1Q1
#Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data (Use iris.csv)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = pd.read_csv(r"Iris.csv")

species_counts = iris_data['Species'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Iris Species (Pie Chart)')
plt.axis('equal')
plt.show()
```

```python
#s3q1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from a CSV file
iris_data = pd.read_csv(r"Iris.csv")


plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_data, x='Species', y='SepalLengthCm', palette='Set1')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Distribution of Sepal Length by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_data, x='Species', y='SepalWidthCm', palette='Set1')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Distribution of Sepal Width by Species')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_data, x='Species', y='PetalWidthCm', palette='Set1')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Distribution of Petal Width by Species')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_data, x='Species', y='PetalLengthCm', palette='Set1')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Distribution of Petal Length by Species')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers
np.random.seed(42)  # Setting seed for reproducibility
random_data = np.random.randint(0, 100, 50)

# Plotting the Line Chart
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.plot(random_data, color='blue', marker='o')
plt.title('Line Chart')
plt.xlabel('Index')
plt.ylabel('Value')

# Plotting the Scatter Plot
plt.subplot(1, 4, 2)
plt.scatter(np.arange(50), random_data, color='green', marker='x')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')

# Plotting the Histogram
plt.subplot(1, 4, 3)
plt.hist(random_data, bins=10, color='orange', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plotting the Box Plot
plt.subplot(1, 4, 4)
plt.boxplot(random_data, vert=False, widths=0.7, patch_artist=True, boxprops=dict(facecolor='pink'))
plt.title('Box Plot')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load the dataset into a DataFrame
df = pd.read_csv('Data.csv')

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# a. One-Hot encoding on the 'Country' column
df_onehot = pd.get_dummies(df, columns=['country'], prefix='country')

# Display the DataFrame after One-Hot encoding
print("\nDataFrame after One-Hot encoding:")
print(df_onehot)

# b. Label encoding on the 'Purchased' column
label_encoder = LabelEncoder()
df['Purchased'] = label_encoder.fit_transform(df['Purchased'])

# Display the DataFrame after Label encoding
print("\nDataFrame after Label encoding on 'Purchased' column:")
print(df)
```
