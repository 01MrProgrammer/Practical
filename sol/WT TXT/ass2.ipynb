```python
import numpy as np
import pandas as pd
data = np.array([30,56,20,30,10,20,90,64,50,60,40,70])
print("\nThe mean of the given data is:")
print(data.mean())
data1 = max(data) - min(data)
print("\nThe range of the given data is:")
print(data1)
q13, q10 = np.percentile(data, [75 ,25])
IQR = q13 - q10
print("\nThe value of IQR for the given data is:")
print(IQR)
```

```python
import pandas as pd
import scipy.stats as s
score={'Name':['Rohit','kartik','umesh','vaishnavi','aakash','shivam','anjali','mugdha','saylee'],
    'graduation percentage': [90,98,78,66,58,83,87,80,90], 'Age':[20,21,21,20,21,20,20,20,21]}
df=pd.DataFrame (score)
print("\nAverage age of students:")
print("Score 2 =", s.tmean(df["Age"]).round(2))
print("\nAverage of graduation percentage:")
print("Score 1 =", s.tmean(df["graduation percentage"]).round(2))
print("\nAll basic statistics of the given data:")
print(df.describe())
```

```python
 
import pandas as pd
import scipy.stats as s
import matplotlib.pyplot as plt
data={'ST':['A','B','C','D','E','F','G','H','I','J'],'DS':[30,25,20,31,29,35,34,33,32,31],
    'OS':[35,34,33,32,31,25,28,29,27,30],'JAVA':[25,28,29,27,30,35,34,32,35,31],
    'TCS':[35,34,32,35,31,35,34,32,32,31],'IP':[35,34,32,32,31,30,25,20,31,29]}
df=pd.DataFrame(data)
df['Geometric Mean']=s.gmean(df.iloc[:,1:10],axis=1)
df['Arithmetic Mean']=s.tmean(df.iloc[:,1:10],axis=1)
df['Harmonic Mean']=s.hmean(df.iloc[:,1:10],axis=1)
print(df)
```
